```python
import shapely.geometry
import pandas as pd
import geopandas as gpd
from keplergl import KeplerGl
from geopandas.tools import geocode
```

```python
from google.colab import output
output.enable_custom_widget_manager()
```

```python
addresses = gpd.read_file('/content/addresses.geojson')
addresses
```

```python
columns = ['addr:city', 'addr:street', 'addr:housenumber']  # список названий столбцов, которые буду объединять
addresses['full_add'] = addresses[columns].apply(', '.join, axis = 1)  # соединяю в единый адрес для получения адреса перед sjoin
addresses.head()
```

```python
buildings = gpd.read_file('/content/buildings_geometries.geojson')
buildings
```

```python
# объединяем 2 таблицы и присваиваем адреса зданиям
buildings_with_add = buildings.sjoin(addresses, how = 'left')
buildings_with_add
```

```python
# визуализируем дома с определенными после объединения адресами
buildings_with_add.loc[buildings_with_add['index_right'].notna()].explore()
```

```python
# тестирую geocode через геокодер osm
geocode('Санкт-Петербург Садовая 45', provider = 'nominatim', user_agent = 'my-application', timeout = None)  
# nominatim - геокодер от osm, user-agent - имя устройства, timeout - время работы функции
```

```python
buildings_without_add = buildings_with_add[buildings_with_add['full_add'].isna()]
buildings_without_add['centr'] = buildings_without_add['geometry'].apply(lambda x: x.centroid)   # определяем центроиды домов для будущего геокодирования
geoc = buildings_without_add[:50]    # отбираю некоторое количество строк для геокодирования (средняя скорость 100 строк в минуту)
```

```python
from geopy.geocoders import Nominatim
geocoder = Nominatim(user_agent = 'my-application', timeout = None)   # устанавливаю параметры геокодирования через nominatim
```

```python
geoc['coord'] = geoc['centr'].apply(lambda z: str(z.y) + ', ' + str(z.x))   
# создаю столбец с необходимым для геокодирования видом написания координат    30.42343, 59.4878264
geoc
```

```python
geoc['geocode_address'] = geoc['coord'].apply(lambda x: geocoder.reverse(x).address)    # обратное геокодирование
# также добавляю .address, так как полный адрес после геокодирования выглядит "5, 10-я Линия, Санкт Петербург, (х, у)", что не позволяет потом провести визуализацию
```

```python
geoc = geoc.drop(['index_right',	'addr:city', 'addr:street', 'addr:housenumber', 'full_add', 'centr', 'coord'], axis = 1)  
#удаляю лишние столбцы (не нужно визуализировать центроиды)
```

```python
# отбираем только те здания, у которых нашли адрес через sjoin
buildings_with_add = buildings_with_add.loc[buildings_with_add['index_right'].notna()]   
```

```python
# 2 разными цветами знаниями с адресами через геокодирование и через объединение таблиц
map = KeplerGl(height = 500)
map.add_data(data = geoc.copy(), name = 'geocode')
map.add_data(data = buildings_with_add.copy(), name = 'address')
map
```
